```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers = pd.read_csv('../data/customers_cleaned.csv', delimiter=';')
customers
```

# Let's start by exploring Income

```python
# Let's target income first, then years with bank

# most correlated features with income
corrmat = customers.corr()
top_corr_features = corrmat.index[abs(corrmat["income"])>0.20]
plt.figure(figsize=(10,10))
g = sns.heatmap(customers[top_corr_features].corr(),annot=True,cmap="RdYlGn_r")

# number of children is somewhat correlated to to income, nbr of children, age
```

```python
sns.boxplot(x='nbr_children', y='income', data=customers)
```

```python
sns.boxplot(x='city_name', y='income', data=customers)
```

```python
sns.boxplot(x='marital_status', y='income', data=customers)
```

# Let's look at years w bank

```python
# most correlated features with income
corrmat = customers.corr()
top_corr_features = corrmat.index[abs(corrmat["years_with_bank"])> .10]
plt.figure(figsize=(10,10))
g = sns.heatmap(customers[top_corr_features].corr(),annot=True,cmap="RdYlGn_r")
```
