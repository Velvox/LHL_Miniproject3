```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import 
scaled_bb_pca = pd.read_csv('../data/scaled_bb_pca.csv', index_col=0)
scaled_bb = scaled_bb_pca.iloc[:,5:9]
scaled_bb
```

## Banking behavior - Savings 
Exploring customers banking behavior with regards to savings

```python
corrmat = scaled_bb.corr()
top_corr_features = corrmat.index[abs(corrmat["total_savings"])>=0]
plt.figure(figsize=(10,10))
g = sns.heatmap(scaled_bb[top_corr_features].corr(),annot=True,cmap="RdYlGn_r")
```
